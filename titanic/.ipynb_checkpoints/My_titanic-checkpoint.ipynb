```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('./train.csv')

train.head()
```

```python
train.info()
```

```python
train.loc[:,'Embarked'].value_counts()
```

```python
train2 = train

train2['Age'] = train2['Age'].fillna((train['Age']).mean())
train2['Cabin'] = train2['Cabin'].fillna('Unknow')
train2['Embarked'] = train2['Embarked'].fillna('S')
```

```python
train2.info()
```

```python
sex_mapDict = {'male':1, 'female':0}
train2['Sex'] = train2['Sex'].map(sex_mapDict)
```

```python
embarked_df = pd.DataFrame()
embarked_df = pd.get_dummies(train2['Embarked'], prefix='Embarked')
tarin2 = pd.concat([train2,embarked_df],axis=1)
tarin2.drop('Embarked',axis=1,inplace=True)
```

```python
pclass_df = pd.DataFrame()
pclass_df = pd.get_dummies(train2['Pclass'], prefix='Pclass')
train2 = pd.concat([tarin2,pclass_df], axis=1)
train2.drop('Pclass',axis=1,inplace=True)
```

```python
def getTitle(name):
    str1 = name.split(',')[1]
    str2 = str1.split('.')[0]
    str3 = str2.strip()
    return str3
```

```python
title_df = pd.DataFrame()
title_df['Title'] = train2['Name'].map(getTitle)
title_df['Title'].unique()
```

```python
title_mapDict={'Mr':'Mr',
               'Mrs':'Mrs',
               'Miss':'Miss',
               'Master':'Master',
               'Dona':'Royalty',
               'Rev':'Officer',
               'Dr':'Officer',
               'Mme':'Mrs',
               'Ms':'Mrs',
               'Major':'Officer',
               'Lady':'Royalty',
               'Sir':'Royalty',
               'Mlle':'Miss',
               'Col':'Officer',
               'Capt':'Officer',
               'the Countess':'Royalty',
               'Jonkheer':'Royalty'    
              }
title_df['Title'] = title_df['Title'].map(title_mapDict)
title_df = pd.get_dummies(title_df['Title'])
train2 = pd.concat([train2,title_df],axis=1)
train2.drop('Name',axis=1,inplace=True)
```

```python
cabin_df = pd.DataFrame()
train2['Cabin'] = train['Cabin'].map(lambda c:c[0])
cabin_df = pd.get_dummies(train2['Cabin'], prefix='Cabin')
train2 = pd.concat([train2,cabin_df], axis=1)
train2.drop('Cabin',axis=1,inplace=True)
```

```python
family_df = pd.DataFrame()
family_df['FamilySize'] = train2['Parch'] + train2['SibSp'] +1
family_df['Family_Single'] = family_df['FamilySize'].map(lambda s :1 if s==1 else 0)
family_df['Family_Small'] = family_df['FamilySize'].map(lambda s :1 if 2<=s<=4 else 0)
family_df['Family_Large'] = family_df['FamilySize'].map(lambda s :1 if 5<=s else 0)
train2 = pd.concat([train2,family_df],axis=1)
```

```python
train2.drop(['PassengerId','Ticket'], axis=1, inplace=True)
train2.head()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(25,18))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(train2.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
corr_df = train2.corr()
corr_df['Survived'].sort_values(ascending = False).round(3)
```

```python
feature_x = pd.concat([title_df,pclass_df,family_df,train2['Fare'],cabin_df,embarked_df,train2['Sex']], axis=1)
feature_x.head()
```

```python
feature_y = train2.Survived
feature_y.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(feature_x, feature_y)
```

```python
model.score(feature_x, feature_y)
```

```python
test = pd.read_csv('./test.csv')

test.info()
```

```python
test['Age'] = test['Age'].fillna((test['Age']).mean())
test['Cabin'] = test['Cabin'].fillna('Unknow')
test['Fare'] = test['Fare'].fillna((test['Fare']).mean())

test.head()
```

```python
test['Sex'] = test['Sex'].map(sex_mapDict)

embarked_df2 = pd.DataFrame()
embarked_df2 = pd.get_dummies(test['Embarked'], prefix='Embarked')
test = pd.concat([test,embarked_df2],axis=1)
test.drop('Embarked',axis=1,inplace=True)

pclass_df2 = pd.DataFrame()
pclass_df2 = pd.get_dummies(test['Pclass'], prefix='Pclass')
test = pd.concat([test,pclass_df2], axis=1)
test.drop('Pclass',axis=1,inplace=True)

title_df2 = pd.DataFrame()
title_df2['Title'] = test['Name'].map(getTitle)
title_df2['Title'] = title_df2['Title'].map(title_mapDict)
title_df2 = pd.get_dummies(title_df2['Title'])
test = pd.concat([test,title_df2],axis=1)
test.drop('Name',axis=1,inplace=True)

cabin_df2 = pd.DataFrame()
test['Cabin'] = test['Cabin'].map(lambda c:c[0])
cabin_df2 = pd.get_dummies(test['Cabin'], prefix='Cabin')
test = pd.concat([test,cabin_df2], axis=1)
test.drop('Cabin',axis=1,inplace=True)

family_df2 = pd.DataFrame()
family_df2['FamilySize'] = test['Parch'] + test['SibSp'] +1
family_df2['Family_Single'] = family_df2['FamilySize'].map(lambda s :1 if s==1 else 0)
family_df2['Family_Small'] = family_df2['FamilySize'].map(lambda s :1 if 2<=s<=4 else 0)
family_df2['Family_Large'] = family_df2['FamilySize'].map(lambda s :1 if 5<=s else 0)
test = pd.concat([test,family_df2],axis=1)

test.drop(['PassengerId','Ticket','Age','SibSp'], axis=1, inplace=True)
test.head()
```

```python
pred = model.predict(test).astype(int)
type(pred)
```

```python
gs = pd.read_csv('./gender_submission.csv')

gs['Survived'] = pred
gs.to_csv ('titanic_pred.csv',index=False)
```
